```python
def print_unique_column(df):
    for column in df:
        if df[column].dtypes == 'object':
            print(f' {column} : {df[column].unique()}')
```

```python
print_unique_column(df1)
```

```python
df1.replace('No internet service', 'No', inplace=True)
df1.replace('No phone service', 'No', inplace=True)
```

```python
df1
```

```python
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn','gender']
for col in yes_no_columns:
    df1[col].replace({'Yes': 1,'No': 0},inplace=True)
```

```python
df1.dtypes
```

```python
for column in df1:
        print(f' {column} : {df1[column].unique()}')
```

```python
df2 = pd.get_dummies(data=df1,columns=['InternetService','Contract','PaymentMethod'])
df2.columns
```

```python
df2.sample(3)
```

```python
df2['gender'].replace({'Female':1,'Male':0},inplace=True)
df2.dtypes
```

```python
col_to_scale =['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df2[col_to_scale] = scaler.fit_transform(df2[col_to_scale])
```

```python
df2
```

```python
for column in df2:
        print(f' {column} : {df2[column].unique()}')
```

```python
X = df2.drop('Churn', axis= 'columns')
y =df2.Churn
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test  = train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
len(X_train.columns)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(20,input_shape = (26,), activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid'),
    
])

model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics =['accuracy'])
model.fit(X_train,y_train, epochs=100)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_predicted = model.predict(X_test)
y_predicted[:5]
```

```python
y_pred = []
for ele in y_predicted:
    if ele<0.5:
        y_pred.append(0)
    else:
        y_pred.append(1)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test,y_pred))
```

```python
import seaborn as sns

cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
